```python
df["Date"].dt.year 
#mention the year of each row
#.dt is a Datetime accessor
```

```python
df["Year"] = df["Date"].dt.year 
df["Month"] = df["Date"].dt.month 

# split the date column into a year column and a month column
```

```python
df
```

```python
df["Day"] = df["Date"].dt.day
df["Weekday"] = df["Date"].dt.day_name()

# split the date column into a day(1-31) column and day_name(weekday) column
```

```python
df
```

```python
df.columns
```

```python
df
```

```python
df["Discounted Price"] = df["Price"] * (1- df["Discount"]/100)
# for each row's price there are discount corresponding that 
# example- Price 100	Discount (%)20	Discounted Price 80
# Price = 72.72 Discount = 5%  Discounted Price 69.084
```

```python
df
```

```python
df["Sell Through Rate"] = df["Units Sold"] / df["Inventory Level"]
# how much of your inventory has been sold.
# 1.0 (100%) → Everything sold (excellent demand)
# 0.5 (50%) → Half inventory sold (moderate)
# 0.1 (10%) → Low sales (slow-moving product)
```

```python
df
```

```python
df["Sell Through Rate"].describe()
# gives statisticsl information
```

```python
df.groupby("Category")["Demand"].agg(["mean","sum","std"]).sort_values(by="sum",ascending=False).reset_index()
# create group of various categories through there mean,sum,std,and sort by sum values in decreasing order
```

```python
df.groupby(["Region","Seasonality"]) ["Demand"].mean()
```

```python
df.groupby("Promotion")["Demand"].mean()
```

```python
df
```

```python
pd.pivot_table(df, values="Demand", index="Month", columns="Category",aggfunc="mean")
```

```python
# histogram plot for demand distribution
sns.histplot(df["Demand"], bins = 20, kde = True)
plt.title("Demand Distribution")
plt.show()
```

```python
# scatter plot for Inventory vs Units Sold
sns.scatterplot(data = df, x= "Inventory Level", y = "Units Sold")
plt.title("Inventory vs Units Sold")
plt.show()
```

```python
# box plot for category vs demand
sns.boxplot(data = df, x = "Category", y = "Demand")
plt.xticks(rotation = 45)
plt.title("Demand by Category")
plt.show()
```

```python
# box plot for weather condition vs demand
sns.boxplot(data = df, x = "Weather Condition", y = "Demand")
plt.xticks(rotation = 45)
plt.title("Demand by Weather Condition")
plt.show()
```

```python
df
```

```python
# monthly seasonality
monthly_demand = df.groupby("Month")["Demand"].mean()
```

```python
monthly_demand
```

```python
# average monthly demand 
monthly_demand.plot(kind = "bar")
plt.title("Average Demand by Month")
plt.show()
```

```python
#overall demand trend
daily_demand = df.groupby("Date")["Demand"].sum()
```

```python
daily_demand
```

```python
# daily demand plot
daily_demand.plot()
plt.title("Total Daily Demand Over Time")
plt.xlabel("Date")
plt.ylabel("Demand")
plt.show()
```

```python
# promotion and pricing diff.
sns.barplot(data = df, x = "Promotion", y = "Demand")
plt.title("Promotion Impact On Demand")
plt.show()
```

```python
# price and demand relationship
sns.scatterplot(data = df, x = "Discounted Price", y = "Demand")
plt.title("Discounted Price vs Demand")
plt.show()
```

```python
# seasonality and the effect on demand
df.groupby("Seasonality")["Demand"].mean().plot(kind = "bar", title = "Demand by Seasonality")
plt.show()
```

```python
# Epidemic and the effect on demand
df.groupby("Epidemic")["Demand"].mean().plot(kind = "bar", title = "Demand by Epidemic")
plt.show()
```
